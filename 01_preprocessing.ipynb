## Preprocessing of Datasets

Normalize data

Handle missing values

Merge datasets to get full GDP data

### Data source

Data from **Kaggle**

GDP data from 1960-2020, **Ozgur Cem Tas**: https://www.kaggle.com/datasets/zgrcemta/world-gdpgdp-gdp-per-capita-and-annual-growths

GDP data from 2020-2025, **Nadiia**: https://www.kaggle.com/datasets/codebynadiia/gdp-per-country-20202025

```python
# imports
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
# load raw data files
data1960 = pd.read_csv("./data/1960-2020.csv")
data1960.head()
```

```python
data2020 = pd.read_csv("./data/2020-2025.csv")
data2020.head()
```

```python
data2020.info()
```

```python
data1960.info()
```

## Reformat to long format

instead of wide format.

```python
data1960 = data1960.melt(id_vars=['Country Name', 'Code'], var_name='Year', value_name='GDP')
data1960.head()
```

```python
data2020 = data2020.melt(id_vars=['Country'], var_name='Year', value_name='GDP')
data2020.head()
```

## Scale datasets to match

1960-> has no rounding

2020-> has been rounded to millions

**Scale both to billions**

```python
# copies of original for long format
data1960_long = data1960.copy()
data2020_long = data2020.copy()
```

```python
# 1. Convert Raw Dollars -> Billions
data1960_long['GDP'] = data1960_long['GDP'] / 1_000_000_000

# 2. Round to 2 decimal places
data1960_long['GDP'] = data1960_long['GDP'].round(2)
data1960_long.head()
```

```python
# Convert Millions to Billions
data2020_long['GDP'] = data2020_long['GDP'] / 1000

# Round to 2 decimals for cleanliness
data2020_long['GDP'] = data2020_long['GDP'].round(2)
data2020_long.head()
```

## Remove unwanted data and clean up

```python
# remove year "Unnamed: 65" 
data1960_long = data1960_long[data1960_long['Year'] != 'Unnamed: 65']       
```

```python
# remove Code column
data1960_long = data1960_long.drop(columns=['Code'])
```

```python
# rename Country Name to Country for consistency
data1960_long = data1960_long.rename(columns={'Country Name': 'Country'})
```

```python
# remove countries from 1960 dataset that are not in 2020 dataset
countries_2020 = set(data2020_long['Country'].unique())
data1960_long = data1960_long[data1960_long['Country'].isin(countries_2020)]
```

## Merge datasets

```python
# merge datasets
full_data = pd.concat([data1960_long, data2020_long], ignore_index=True)
```

```python
print(full_data.head())
print(full_data.tail())
```

## test

visualization

```python
# unique countries
full_data['Country'].unique()
```

```python
# finland data
full_data[full_data['Country'] == 'Sweden']
```

```python
# basic line plot of GDP over time for a few countries
countries_of_interest = ['United States', 'China', 'Finland', 'Germany', 'Sweden']
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = full_data[full_data['Country'] == country]
    plt.plot(country_data['Year'], country_data['GDP'], label=country)

plt.title('GDP Over Time (1960-2020)')
plt.xlabel('Year')
plt.ylabel('GDP (Billions of USD)')
plt.legend()
plt.xticks(rotation=65)
plt.show()
```

```python
country = 'Israel'
# highlight missing values
plt.figure(figsize=(12, 6))
country_data = full_data[full_data['Country'] == country]
plt.plot(country_data['Year'], country_data['GDP'], label=country)
# for missing data points, we can add markers
# identify and plot markers for NaN values at y=0 (or bottom of chart)
missing = country_data[country_data['GDP'].isna()]
plt.scatter(missing['Year'], [0] * len(missing), 
            color='red', marker='x', label=f'{country} (Missing)')

plt.title('GDP Over Time (1960-2025) - Highlighting Missing Data')
plt.xlabel('Year')
plt.ylabel('GDP (Billions of USD)')
plt.legend()
plt.xticks(rotation=65)
plt.show()
```
